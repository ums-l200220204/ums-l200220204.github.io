```python
# Step 1: Import libraries
import pandas as pd
import re

# Step 2: Upload file WhatsApp (.txt)
from google.colab import files
uploaded = files.upload()  # Pilih file WhatsApp export .txt

# Step 3: Baca file WhatsApp
file_name = list(uploaded.keys())[0]  # Ambil nama file
with open(file_name, 'r', encoding='utf-8') as file:
    whatsapp_data = file.readlines()

# Step 4: Parsing data WhatsApp
parsed_data = []

# Regex disesuaikan untuk menangani berbagai format tanggal/waktu
regex_pattern = r'(\d{2}[/.-]\d{2}[/.-]\d{2,4},? \d{2}[:.]\d{2}) - (.*?): (.*)'

for line in whatsapp_data:
    # Cek apakah formatnya sesuai dengan pola tanggal + waktu di WhatsApp
    match = re.match(regex_pattern, line)
    if match:
        date_time = match.group(1)
        sender = match.group(2)
        message = match.group(3)
        parsed_data.append({'date_time': date_time, 'sender': sender, 'message': message})
    else:
        # Tambahkan pesan yang terpotong ke baris sebelumnya
        if parsed_data:
            parsed_data[-1]['message'] += f" {line.strip()}"

# Step 5: Buat DataFrame dari hasil parsing
df = pd.DataFrame(parsed_data)

# Step 6: Konversi kolom date_time ke tipe datetime
# Deteksi format waktu secara otomatis
df['date_time'] = pd.to_datetime(df['date_time'], dayfirst=True, errors='coerce')

# Step 7: Tampilkan DataFrame
print("Preview DataFrame:")
print(df.head())

# Step 8: Save DataFrame ke file CSV (opsional)
df.to_csv('whatsapp_parsed.csv', index=False)
print("File 'whatsapp_parsed.csv' telah disimpan.")
```

```python
import pandas as pd
import re

# Fungsi untuk membersihkan teks
def clean_text(text):
    if isinstance(text, str):
        # Hanya mempertahankan angka, huruf, dan tanda baca-tulis umum
        return re.sub(r'[^a-zA-Z0-9.,:;!?()\-"\'"\" ]', '', text)
    return text

# Baca file CSV
file_path = 'whatsapp_parsed.csv'
# Specify the separator as a comma ','
data = pd.read_csv(file_path, sep=',') # Changed sep to ','

# Bersihkan setiap elemen dalam DataFrame
data_cleaned = data.applymap(clean_text)

# Simpan hasil ke file baru
output_path = 'data_group_cleaned.csv'  # Simpan di direktori kerja saat ini
data_cleaned.to_csv(output_path, index=False)
print(f"Data cleaned and saved to {output_path}")
```
